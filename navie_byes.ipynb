```python
#Text Preprocessing
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('vader_lexicon')
nltk.download('averaged_perceptron_tagger')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
print(stop_words)
```

```python
print(lemmatizer)
```

```python
def preprocess_text(text):
      # Remove punctuation and numbers
      text = re.sub(r'[^\w\s]', '', text)
      text = re.sub(r'\d+', '', text)
      # Convert to lowercase
      text = text.lower()
      # Remove stopwords
      stop_words = set(stopwords.words('english'))
      text = ' '.join([word for word in text.split() if word not in stop_words])
      return text
      # Apply lemmatization
      text = ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
      return text
      # Apply POS tagging
      pos_tags = nltk.pos_tag(nltk.word_tokenize(text))
      return pos_tags
      # Apply NER
      ner_tags = nltk.ne_chunk(pos_tags)
      return ner_tags
      # Apply sentiment analysis
      sentiment = TextBlob(text).sentiment
      return sentiment
      # Apply topic modeling
      topics = LatentDirichletAllocation(n_components=5).fit_transform(text)
      return topics
      # Apply text summarization
      summary = summarize(text)
      return summary
      # Apply text classification
      classifier = MultinomialNB().fit(text)
      return classifier
      # Apply text generation
      generator = pipeline('text-generation', model='gpt2')
      return generator
      # Apply text translation
      translator = Translator()
      translation = translator.translate(text)
      return translation
      # Apply text similarity
      similarity = cosine_similarity(text)
      return similarity
      # Apply text clustering
      clusterer = KMeans(n_clusters=5).fit(text)
      return clusterer
      # Apply text compression
      compressor = LempelZivMarkovChain(text)
      return compressor
      # Apply text encryption
      cipher = AES.new('This is a key123', AES.MODE_CBC, 'This is an IV456')
      encrypted = cipher.encrypt(text)
      return encrypted
      # Apply text decryption
      cipher = AES.new('This is a key123', AES.MODE_CBC, 'This is an IV456')
      decrypted = cipher.decrypt(encrypted)
      return decrypted
      # Apply text compression
      compressor = gzip.compress(text)
      return compressor
      # Apply text decompression
      decompressor = gzip.decompress(compressor)
      return decompressor
      # Apply text compression
      compressor = bz2.compress(text)
      return compressor
      # Apply text decompression
      decompressor = bz2.decompress(compressor)
      return decompressor
      # Apply text compression
      compressor = lzma.compress(text)
      return compressor
      # Apply text decompression
      decompressor = lzma.decompress(compressor)
      return decompressor
      # Apply text compression
      compressor = zlib.compress(text)
      return compressor
      # Apply text decompression
      decompressor = zlib.decompress(compressor)
      return decompressor
      # Apply text compression
      compressor = snappy.compress(text)
      return compressor
      # Apply text decompression
      decompressor = snappy.decompress(compressor)
      return decompressor
      # Tokenize and remove stopwords
      tokens = nltk.word_tokenize(text)
      tokens = [token for token in tokens if token not in stop_words]
      return tokens




```

```python
# Apply preprocessing
df['Data'] = df['Data'].apply(preprocess_text)
df.head()
```

```python
#Feature Extraction
#Use TF-IDF to convert the cleaned text data into numerical format.
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
x = vectorizer.fit_transform(df['Data'])
print(x)

```

```python
y=df['Labels']
print(y)
```

```python
#2. Naive Bayes Model for Text Classification
#Splitting the Data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
#Naive Bayes Model for Text Classification
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred)
confusion_matrix(y_test,y_pred)
print(classification_report(y_test,y_pred))
```

```python
#Sentiment Analysis
#Performing Sentiment Analysis
from textblob import TextBlob
df['sentiment']=df['Data'].apply(lambda x:TextBlob(x).sentiment.polarity)
df.head()

```

```python
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'negative'

df['sentiment'] = df['Data'].apply(get_sentiment)
df
```
